```python
import matplotlib.pyplot as plt # for data visualization purposes
import pandas as pd
import numpy as np
from sklearn import datasets # for importing the iris datasets
from sklearn.decomposition import PCA
```

```python
breast = datasets.load_breast_cancer()
breast_data = breast.data # breast_data has access to all the features (for e.g. symmetry error)
# of the dataset that can be used to classify the sample
breast_labels = breast.target # breast.target would have species
# itself so what data is being classified as
breast_data.shape
```

```python
breast_labels.shape
```

```python
labels = np.reshape(breast_labels, (569, 1)) # reshaping breast label from (569, ) to (569, 1)
final_breast_data = np.concatenate([breast_data, labels], axis =1) # joining the reshaped label to the original dataset
# in order to create a dataframe containing both the data and labels
final_breast_data.shape
```

```python
breast_dataset = pd.DataFrame(final_breast_data) # creating dataframe
# checking features
features = breast.feature_names
features # as can be seen in the output below, label is missing. It will be added soon
```

```python
breast_dataset.head()
```

```python
features_labels = np.append(features, 'label') # adding in the feature label
breast_dataset.columns = features_labels # embedding column names into table
breast_dataset.head()
```

```python
# replacing the 0 and 1 values in label column with benign and malignant values
breast_dataset['label'].replace(0, 'Benign', inplace=True)
breast_dataset['label'].replace(1,'Malignant', inplace=True)
breast_dataset.head()
```

#### Implementing PCA

```python
# normalizing data before applying PCA to it
from sklearn.preprocessing import StandardScaler # module used to normalize data
x = breast_dataset.loc[:, features].values # selecting features from the breast_dataset
# (NOT the dataframe that was constructed earlier)
x = StandardScaler().fit_transform(x) # scaling the features by applying fit_transform on the feature data
x.shape
```
